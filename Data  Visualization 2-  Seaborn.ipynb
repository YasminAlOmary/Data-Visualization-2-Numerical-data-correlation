```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df=sns.load_dataset('tips') 
num=df._get_numeric_data() 
```

```python
num
```

```python
# to find correlation between data (positive -Negative )
num.corr()
```

```python
sns.regplot(df['tip'],df['total_bill'])
```

```python
# fit_reg=False to remove the line in chart
sns.regplot(df['tip'],df['total_bill'],fit_reg=False)
```

```python
# scatter_kws={'alpha':0.3} to appear the transparency and density of the data
sns.regplot(df['tip'],df['total_bill'],fit_reg=False ,scatter_kws={'alpha':0.3})
```

```python
sns.regplot(df['tip'],df['total_bill'],fit_reg=False ,scatter_kws={'alpha':0.7})
```

```python
#lmplot to plot data and draw regression model
#hue='gender'to appear the distribution of gender 
df.rename(columns={'sex':'gender'},inplace=True) 
sns.lmplot(x='tip',y='total_bill',data=df ,fit_reg=False ,scatter_kws={'alpha':0.15} ,hue='gender')    
```

```python
# plt.title:  to added title to the chart
sns.lmplot(x='tip',y='total_bill',data=df ,fit_reg=False ,scatter_kws={'alpha':0.15} ,hue='gender')   
plt.title ('tip and Total bill relation ')
```

```python
# plt.axvline : to draw mean of tips with color red in chart.
sns.lmplot(x='tip',y='total_bill',data=df ,fit_reg=False ,scatter_kws={'alpha':0.15} ,hue='gender')   
plt.title ('tip and Total bill relation ')
plt.axvline(df['tip'].mean(),c='r') 
```

```python
sns.lmplot(x='tip',y='total_bill',data=df ,fit_reg=False ,scatter_kws={'alpha':0.15} ,hue='gender')   
plt.title ('tip and Total bill relation ')
plt.axvline(df['tip'].mean(),c='b') 
```

```python
# sns.jointplot: to appear  peak and frequency of the distribution  
sns.jointplot(df['tip'],df['total_bill'],kind='kde') 
```

```python
# stats.pearsonr : to find correlation and  probability between the two column
stats.pearsonr(df['tip'],df['total_bill']) 
```

```python
stats.pearsonr(df['tip'],df['size']) 
```

```python
# annotate(stats.pearsonr): to draw pearsonr
sns.jointplot(df['tip'],df['total_bill'],kind='kde').annotate(stats.pearsonr)
```

```python
# FacetGrid : to view grouped between two  data
plt.figure(figsize=(15,10))
sns.FacetGrid (df, col='gender',row='smoker' ).map(sns.regplot,'tip','total_bill')   
```

```python
plt.figure(figsize=(15,10))
sns.FacetGrid (df, col='day',row='smoker' ).map(sns.regplot,'tip','total_bill')  
```

```python
plt.figure(figsize=(15,10))
sns.FacetGrid (df, col='day',row='smoker' ).map(sns.regplot,'tip','size')  
```

```python
plt.figure(figsize=(15,10))
sns.FacetGrid (df, col='gender',row='smoker' ).map(sns.boxplot,'tip')
```

```python
#tip and day grouped by box plot
plt.figure(figsize=(15,10))
sns.FacetGrid (df, col='gender',row='smoker' ).map(sns.boxplot,'day','tip')  
```
